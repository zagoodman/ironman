```python
# Multiple finishes

# Check uniqueness of bib numbers within a race
print(len(df))
print(len(df.loc[:,['raceid', 'bib_n']].drop_duplicates()))

# N - should be 1. This is listings per athlete-race
df['N'] = df.groupby(['raceid', 'athleteid']).athleteid.transform('count')
print(df.N.value_counts())
```

```python
# entry data pulled from races may help fix duplicates issues

dfc = pd.read_csv('../data/im_full_with_country.csv')

# keep only necessary vars and observations (2010 - 2019)
dfc = dfc[dfc.raceyear >= 2010]
dfc = dfc.loc[dfc.finish != "DNS"]
dropvars = ['swimrank', 'swimrankdiv', 'swimrankgender', 
            'bikerank', 'bikerankdiv', 'bikerankgender',
            'runrank', 'runrankdiv', 'runrankgender', 
            'trans1rank', 'trans1rankdiv', 'trans1rankgender', 
            'trans2rank', 'trans2rankdiv', 'trans2rankgender', 'half']
dfc.drop(dropvars, 1, inplace=True)
dfc = dfc.drop_duplicates()
dfc.sort_values(['athleteid', 'raceyear', 'racename', 'division', 'overalltime'], inplace=True)

# fix missings
for c in [x for x in dfc.columns if x not in ['athleteid']]:
    dfc.loc[dfc[c] == 99999, c] = np.NaN
display(dfc.isnull().sum())

dfc.head(1)
```

```python
# Check multiple finishes for dfc

# N - should be 1. This ensures uniqueness of bib numbers within a race
dfc['N'] = dfc.groupby(['raceid', 'bib_n']).bib_n.transform('count')
print(dfc.N.value_counts())

# Fix duplicated entry
dfc.drop(index=dfc.loc[(dfc.N == 2) & (dfc.athleteid.isnull())].index, inplace=True)
#display(dfc.loc[dfc.N == 2])

# Checm uniqueness of bib numbers within a race
print(len(dfc))
print(len(dfc[['raceid', 'bib_n']].drop_duplicates()))

# N - should be 1. This is listings per athlete-race
dfc['N'] = dfc.groupby(['raceid', 'athleteid']).athleteid.transform('count')
print(dfc.N.value_counts())
```

```python
# despite having unique bib_n, some athleteids have duplicate entries. 
# Why? Because:
# 1. athleteid is assigned at reg, and some family members reg together.
# 2. athletes assigned multiple bib_n and only counted once
# 3. some athletes listed in multiple categories

# Goal: assign unique athleteid such that raceid-athleteid uniquely identifies observations
# To do this, we will add 1e7 to athleteids that are redundant
# Importantly, we want the true account holder to stay with the original athleteid
# We will match the true account holder using the name on the account or gender
# If there are different divisions but same name/gender, we will give the older priority


# Steps:
# 1) identify accounts that need spliting
# 2) separate further accounts that can be split 1) by gender, 2) by name and not gender, 3) by division and not name or gender
# 3) among accounts that cannot be split (constant in race-id-name-gender-division), assign the faster account priority

# Notes: for 2), change the athleteid for those in the main df, in case some "guests" raced solo, but only for gender and name
#  For 2.3 and 3, do NOT apply changes to main DF as later we will look for age group issues.
# Some entries are errors (e.g. pros listed as amateurs), but in these steps we will not delete entries.
```

```python
# 1) split sample
dfsub = dfc.loc[dfc.N > 1]
#dfc.drop(index=dfsub.index, inplace=True)

# merge account name onto dfsub
dfname = df.loc[:, ['athleteid', 'athletename']].drop_duplicates()
dfname.columns = ['athleteid', 'account_name']
dfsub = dfsub.merge(dfname, on='athleteid', how='left')
dfsub.head()

# Return dfsub back to dfc
#dfc = pd.concat([dfc, dfsub])
```

```python
# 2.1 Split accounts by gender

# calc num genders per athleteid in dfsub
dfsub['Ngen'] = dfsub.groupby('athleteid').gender.transform('nunique')
print(dfsub.Ngen.value_counts())

# get subset with Ngen > 1
dfgen = dfsub.loc[dfsub.Ngen > 1].sort_values(['athleteid', 'raceid', 'gender']).reset_index(drop=True)

# assign account_gender using account_name
dfgen['account_gender'] = dfgen.loc[dfgen.athletename == dfgen.account_name, 'gender']
dfgen.account_gender = dfgen.groupby('athleteid').account_gender.fillna(method='ffill')
dfgen.account_gender = dfgen.groupby('athleteid').account_gender.fillna(method='bfill')
print(dfgen.account_gender.value_counts())

# loop through and manually label where names don't match
for i in range(int(len(dfgen)/2)):
    dftmp = dfgen.loc[i*2:i*2+1, ['athletename', 'account_name', 'athleteid', 'gender', 'account_gender']].reset_index(drop=True)
    if dftmp.account_gender.notnull().sum() > 0:
        continue
    display(dftmp)
    idx = int(input("Enter account index (0 or 1): "))
    dfgen.loc[i*2:i*2+1, 'account_gender'] = "Female" if idx == 0 else "Male" if idx == 1 else "error"
```

```python
# Verify that account_gender defined for everyone in dfgen
print(dfgen.account_gender.isnull().sum())

# Loop through dfgen replacing athleteids in dfc and df

def replace_by_gender(id, account_gender):
    dfc.loc[(dfc.athleteid == id) & (dfc.gender != account_gender), 'athleteid'] = id + 1e7
    df.loc[(df.athleteid == id) & (df.gender != account_gender), 'athleteid'] = id + 1e7

dftmp = dfgen.loc[:, ['athleteid', 'account_gender']].drop_duplicates().reset_index(drop=True)
for i in range(int(len(dftmp))):
    replace_by_gender(dftmp.loc[i, 'athleteid'], dftmp.loc[i, 'account_gender'])
```

```python
# recalc N and get new dfsub
dfc['N'] = dfc.groupby(['raceid', 'athleteid']).athleteid.transform('count')
print(dfc.N.value_counts())
dfsub = dfc.loc[dfc.N > 1]

# Step 2.2: split by name, division, and/or time and not gender
# We'll compare each name to the account name and assign the account holder

# merge account name onto dfsub
dfname = df.loc[:, ['athleteid', 'athletename']].drop_duplicates()
dfname.columns = ['athleteid', 'account_name']
dfsub = dfsub.merge(dfname, on='athleteid', how='left')
dfsub.head(10)

# export for offline
dfsub.to_csv('../data/dfsub_20201021.csv', index=False)
```

```python
# name cleaning

# If comma, move everything after the comma to front of name
# lowercase everything
# remove special characters

def name_clean(n):
    """
    Takes a name of the form Last, First Middle
    And returns first middle last. Note that Last can
    have multiple parts, and there may not be a middle name.
    Returned name is lowercase.
    Removes common endings to names (like Jr, Sr, etc.)
    """
    # check if string
    if str(n) != n:
        return
    
    # lowercase
    n = n.lower()
    
    # remove prefixes and suffixes
    remove_words = [' jr.,', ' jr.', ' jr', ' dr.,', ' dr.', ' dr', ' sr.', ' sr']
    for j in remove_words:
        n = n.replace(j, "")
    
    # remove special characters (keep only letters)
    n = ''.join(e for e in n if e.isalpha() or e.isspace() or e == ",")
    
    # rearrange if any commas found
    n_list = n.split(", ")
    
    # if no commas found, return n
    if len(n_list) == 1:
        return n
    
    # If comma found, flip order
    name = n_list[1] + " " + n_list[0]
    return name


# special cases - remove double commas (only six)
def name_special_case(id, bib, name):
    dfm.loc[(dfm.athleteid == id) & (dfm.bib_n == bib), 'athletename'] = name

name_special_case(131572, 452, 'Kim Lun Lui')
name_special_case(148002, 1114, 'Anthony Kiang Meng Tan')
name_special_case(227395, 2567, 'Juan Marquez')
name_special_case(518176, 1335, 'Ruben De La Pena Gallego')
name_special_case(540638, 1448, 'Eric Lo Ying Wai')
name_special_case(540720, 1503, 'Maurice Hung Wen-Hsiang')
```
